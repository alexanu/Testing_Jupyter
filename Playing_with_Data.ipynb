Below code get 1Y daily quotes for a random stock

```python
ticker = random.choice(["FB","AAPL","NFLX","GOOG","BA","GS","BABA","TSLA"])
# selected_ticker = yf.Ticker(ticker)
hist = yf.Ticker(ticker).history(period="1y")

plt.plot(hist.index,hist['Close'])
plt.xlabel('date')
plt.ylabel('$ price')
plt.title(ticker+" stock price")
```
